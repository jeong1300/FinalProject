```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import matplotlib.font_manager as fm
font_name = fm.FontProperties(fname="C:/Windows/Fonts/malgun.ttf").get_name()
plt.rc('font', family=font_name)

mpl.rcParams["axes.unicode_minus"] = False
```

```python
# 넷플릭스 자료 불러오기
netflix = pd.read_csv("C:/Users/User_/jeong/data_plus_star.csv")
netflix
```

```python
# 복사본 제작
netflix1 = netflix
```

```python
# rating 종류
netflix1['rating'].unique()
```

```python
# rating 종류개수
netflix1['rating'].value_counts()
```

```python
# rating 결측치 개수
netflix1['rating'].isnull().sum()
```

```python
# rating 결측치 6(전체관람가)로 대체
netflix1['rating'] = netflix['rating'].fillna(6)
```

```python
# rating 범주 분류
netflix1['rating'] = netflix['rating'].replace(['TV-MA', 'R', 'NC-17'], 0)
netflix1['rating'] = netflix['rating'].replace(['TV-14', 'PG-13'], 1)
netflix1['rating'] = netflix['rating'].replace(['TV-PG', 'TV-G', 'PG'], 2)
netflix1['rating'] = netflix['rating'].replace(['TV-Y7-FV', 'TV-Y7'], 3)
netflix1['rating'] = netflix['rating'].replace(['TV-Y'], 4)
netflix1['rating'] = netflix['rating'].replace(['NR', 'UR'], 5)
netflix1['rating'] = netflix['rating'].replace(['G'], 6)
```

```python
# rating 갯수 
netflix1["rating"].unique()
netflix1['rating'].value_counts()
```
